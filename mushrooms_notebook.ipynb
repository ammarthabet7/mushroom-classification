# data preprocessing

```python
#droping those highly skewed columns
df.drop(['bruises','gill-attachment','gill-spacing','gill-size','stalk-shape','veil-color','ring-number'],axis =1,inplace=True)
```

**Encoding data**:it is a very important step because the dataset features are all categorys, so we must choose the suitable way of encoding.

Now we will try to give our models a **numeric representation** for every category with a small number of columns but with an encoding that will put similar categories close to each other. **The easiest way to do it is replace every category with the number of times that we saw it in the dataset.**

```python
df=df.copy()
columns=list(df.columns)
columns.pop(0)
for i in columns:
    if(df[i].dtype=='object'):
        df[i]=df[i].astype('category')
        counts=df[i].value_counts()
        counts=counts.sort_index()
        counts += np.random.rand(len(counts))/1000
        df[i].cat.categories=counts
```

```python
df['class'] = df['class'].replace("p", "0" )
df['class'] = df['class'].replace("e", "1" )
```

```python
df.head()
```

**Important note**: its not logical to search for **outliers** or look at **coorelations** because the dataset features all are **categorys** so, thoose numbers are only from **encoding the data** (numbers reprsenting the name of the category not its value)

# modeling

```python
X=df.drop("class",axis=1)
y=df['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)
```

```python
#@title decision tree
# import the regressor
from sklearn.tree import DecisionTreeClassifier

from sklearn.tree import plot_tree

plt.figure(figsize=(15,8))
clf = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0).fit(X, y)
plot_tree(clf.fit(X_train,y_train), filled=True)
plt.title("Decision tree model")
plt.show()

```

```python
from sklearn.metrics import accuracy_score

#Predict the values 
y_pred_gini = clf.predict(X_test)

#Predict the value using X train for accuracy comparision 
y_pred_train_gini = clf.predict(X_train)
```

```python
#Determine the accuracy score
print('Model accuracy score with criterion gini index: {0:0.4f}'. format(accuracy_score(y_test, y_pred_gini)))
#Accuracy Score for training set
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_gini)))
```
